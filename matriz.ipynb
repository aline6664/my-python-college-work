### 2. Faça um programa que leia uma matriz 2 x 2, que representa o salário de 4 funcionários, calcule e mostre a soma total de todos os salários/elementos que será o montante pago pela empresa a esses funcionários.

### 4. Faça um programa que leia números inteiros w e p e os elementos de uma matriz A de números inteiros de dimensão w x p e conte o número de elementos que são iguais a zero. Observação w representa a quantidade de linhas e p as colunas


```
A = []
w = int(input('Informe a quantidade de linhas: '))
p = int(input('Informe a quantidade de colunas: '))
cont = 0

for linha in range(w):
    vetor_linha = []
    for coluna in range(p):
        vetor_linha.append(int(input(f'Digite um elemento [{linha},{coluna}]: ')))
        A.append(vetor_linha)

for linha in range(w):
    for coluna in range(p):
        if A[linha][coluna] == 0:
            cont += 1
print('Número de elementos iguais a zero (0):',cont)
```

### 5. Faça um programa que leia uma matriz 3 x 2, que representa preços de produtos, crie OUTRA matriz que armazene todos os preços com 7% de aumento.

```
preco = []
for linha in range(3):
    vetor_linha = []
    for coluna in range(2):
        vetor_linha.append(float(input(f'Digite o preço[{linha}][{coluna}] R$: ')))
    preco.append(vetor_linha)

aumento = []
for linha in range(len(preco)):
    vetor_linha = []
    for coluna in range(len(preco[0])): # -> qtde de colunas que tem na linha 0, contando os elementos inseridos
        vetor_linha.append(preco[linha][coluna] * 1.07) # ou preco[linha][coluna] + preco[linha][coluna] * 7/100
    aumento.append(vetor_linha)

print()
print('Apresentação da matriz preço........')
for linha in range(3):
    for coluna in range(2):
        print(preco[linha][coluna],end='\t')
    print()

print()
print('Apresentação da matriz aumento........')
for linha in range(3):
    for coluna in range(2):
        print(f'{aumento[linha][coluna]:.2f}',end='\t')
    print()
```

### 6. Faça um programa que leia: 
###* um vetor com oito posições com os nomes das lojas; 
###* um outro vetor com quatro posições com os nomes dos produtos;
###* uma matriz (8 x 4) com os preços de todos os produtos em cada loja. 
###O programa deve mostrar todas as relações/print (nome da loja - nome do produto e preço), nas quais o preço não ultrapasse R$ 120,00.


```
i_loja = 1
i_prod = 1

loja = []
while i_loja <= 4:
    loja.append(input(f'Nome da {i_loja}º loja: '))
    i_loja = i_loja + 1
print()

produto = []
while i_prod <= 3:
    produto.append(input(f'Nome do {i_prod}º produto: '))
    i_prod = i_prod + 1
print()

print(loja)
print(produto)

print()
matriz = []
for linha in range(len(loja)):
    vetor_linha = []
    for coluna in range(len(produto)):
        vetor_linha.append(float(input(f'Preço do(a) {produto[coluna]} na loja {loja[linha]}: ',)))
    matriz.append(vetor_linha)

print()
for linha in range(len(loja)):
    for coluna in range(len(produto)):
        if matriz[linha][coluna] < 120:
            print('Loja:',loja[linha],'- Produto:',produto[coluna],'- Preço: R$',matriz[linha][coluna])
```

###7. Faça um programa que leia uma matriz 4 x 4. 
###* Apresente os valores/elementos e seus respectivos índices (linhas e colunas) que estão na diagonal principal.
###* Calcule e apresente a soma dos elementos acima da diagonal principal, que estão em negrito e com a cor de fundo amarela.

image.png


```
[0,1][0,2][0,3][1,2][1,3][2,3]
```

```
matriz = []

for linha in range(4):
    vetor_linha = []
    for coluna in range(4):
        vetor_linha.append(int(input(f'Valor[{linha}][{coluna}]: ')))
    matriz.append(vetor_linha)

print()
print('Valores/elementos e seus respectivos índices na diagonal principal')
for linha in range(4):
    for coluna in range(4):
        if [linha] == [coluna]:
            print(matriz[linha][coluna], end='\t')
            print(f'[{linha}][{coluna}]')
print()
print('Soma dos elementos acima da diagonal principal')
soma = 0
acima_diagonal = []

for linha in range(4):
    for coluna in range(4):
        if [coluna] > [linha]:
            soma = soma + matriz[linha][coluna]
            acima_diagonal.append(matriz[linha][coluna])
print(acima_diagonal,'=',soma)
```
