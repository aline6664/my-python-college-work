###1. Construir um algoritmo para calcular e apresentar o total de salários pagos de funcionários, mas não é informado a quantidade de pessoas, então use como critério de parada (condição da estrutura de repetição, digitar zero no salário para sair.


```
print('Programa para somar salário maiores que zero.')
total = 0
salario = float(input('Informe o salário...: '))
while salario > 0: # <- Enquanto salário estiver acima de 0, ele será aceito infinitamente. 
    total = total + salario # Por while estar definido como maior que zero, qualquer valor negativo automaticamente irá parar a repetição.
    salario = float(input('Informe o salário: '))
print(f'Total R$ {total:.2f}')
```

```
# GAMBIARRA ...
total = 0
salario = 1
while salario > 0:
    salario = float(input('Informe o salário: '))
    if salario > 0:
        total = total + salario
print(f'Total R$ {total:.2f}')
```

###2. Faça um programa que receba a altura de 5 pessoas. Encontre e apresente a altura da pessoa mais alta e da mais baixa e seus respectivos nomes.

```
for pessoas in range(5):
    altura = float(input('Informe a altura: '))
    nome = input('Informe o nome: ')
    # dica de implementação, inicializar as variáveis de menor e maior
    # valor dentro do laço de repetição, quando você souber, quantas
    # vezes repete
    if pessoas == 0:
        alto = altura
        nomea = nome
        baixo = altura
        nomeb = nome
    if altura >= alto:
        alto = altura
        nomea = nome
    elif altura <= baixo:
        baixo = altura
        nomeb = nome
print(nomea,'é mais alto, com',alto,'metros')
print(nomeb,'é mais baixo, com',baixo,'metros')
```

###3. Faça um programa que receba a altura de várias pessoas. Encontre e apresente a altura da pessoa mais alta e da mais baixa e seus respectivos nomes. Para encerrar a entrada de dados, zero na altura, mas esta não poderá ser considerada como resposta da altura da pessoa mais baixa.


```
alto = 0
baixo = 10 # <- Valor alto qualquer, é só uma inicialização.
altura = float(input('Informe a altura? ')) # <- Input de inicialização
while altura > 0:
    nome = input('Informe o nome: ') # <- Colocados no inicial e final devido a repetição
    if altura >= alto:
        alto = altura
        nomea = nome
    elif altura <= baixo:
        baixo = altura
        nomeb = nome
    altura = float(input('Informe a altura: '))

print(nomea,'é mais alto, com',alto,'metros')
print(nomeb,'é mais baixo, com',baixo,'metros')
```

###4. Faça um programa que receba a idade e a altura de várias pessoas. Calcule e exiba a média das alturas das pessoas com mais de 20 anos. Para encerrar a entrada de dados, digite uma idade negativa ou igual a zero.




```
soma_alturas = 0
qtde_pessoas = 0
idade = int(input('Digite a idade...: ')) # <- fica fora da repetição apenas a condição da entrada de dados. Nesse caso, a idade.
while idade > 0:
    altura = float(input('Digite a altura: '))
    if idade > 20:
        soma_alturas = soma_alturas + altura
        qtde_pessoas = qtde_pessoas + 1
    idade = int(input('Digite a idade...: '))

if qtde_pessoas > 0: # <- regra matemática, não pode dividir por 0. Não sabemos se vai ter pessoas que atendem ao critério desejado.
    media = soma_alturas / qtde_pessoas
    print(f'A média de alturas é {media:.2f}')
```

###5. Construir um algoritmo para calcular e apresentar a idade atual de algumas pessoas em relação ao ano atual, mas não é informado a quantidade de pessoas, então use como critério de parada (condição da estrutura de repetição, digitar zero no ano de nascimento para sair.

```
ano_atual = 2023
idade = 0
ano_nascimento = int(input('Informe seu ano de nascimento: '))
while ano_nascimento > 0:
    idade = ano_atual - ano_nascimento
    print('Você tem',idade,'anos')
    ano_nascimento = int(input('Informe seu ano de nascimento: '))
```

###6. Faça um programa que receba um conjunto de valores inteiros, calcule e exiba o maior e o menor valor do conjunto.


*   Para encerrar a entrada de dados, deve ser digitado o valor zero;
*   Para valores negativos, deve ser enviada uma mensagem;
*   Esses valores (zero e negativos) não entrarão na lógica de encontrar o maior e o menor valor.



```
maior_valor = 0
menor_valor = 10
valor = int(input('Digite um valor inteiro: '))
while valor != 0:
    if valor > 0:
        if valor >= maior_valor:
            maior_valor = valor
        elif valor <= menor_valor:
            menor_valor = valor
    else:
        print('Erro 101: valor negativo não válido.')
           
    valor = int(input('Digite um valor inteiro... '))

print('O maior valor é',maior_valor,'e o menor valor é',menor_valor,':)')
```

###7. No final do ano muitas pessoas compram presentes. Faça um programa que registre alguns dados das pessoas, usando como critério de parada a letra ‘n’, para a pergunta “Deseja cadastrar outro (‘s’/’n’)?”, para identificar o perfil dos compradores numa loja de roupas e apresente como resultado a:
a)   Quantidade de mulheres e de homens;

b)   Quantidade de mulheres e de homens abaixo e acima de 18 anos.

```
# item a) 2 contadores
# item b) 4 contadores

# a)
qtde_homens = 0
qtde_mulheres = 0

cadastro = input('Deseja cadastrar (s/n)? ')
while cadastro == 's' or cadastro == 'S':
    sexo = input('Sexo (m/f): ')
    if sexo == 'm' or sexo == 'M':
        qtde_homens = qtde_homens + 1
    elif sexo == 'f' or sexo == 'F':
        qtde_mulheres = qtde_mulheres + 1
    else:
        print('Input inválido. Por favor, digite (s/n) para confirmar.')
    cadastro = input('Deseja cadastrar (s/n)? ')

print('Quantidade de homens:',qtde_homens)
print('Quantidade de mulheres:',qtde_mulheres)

if cadastro == 'n' or cadastro == 'N':
    print()
else:
    print('Input inválido. Por favor, digite (s/n) para confirmar!')
```

```
# b)
qtde_homens_mais_18 = 0
qtde_homens_menos_18 = 0
qtde_mulheres_mais_18 = 0
qtde_mulheres_menos_18 = 0


cadastro = input('Deseja cadastrar (s/n)? ')
while cadastro == 's' or cadastro == 'S':
    sexo = input('Sexo (m/f): ')
    idade = int(input('Idade: '))

    if sexo == 'm' or sexo == 'M':
        qtde_homens = qtde_homens + 1
        if idade < 18 and idade > 0:
            qtde_homens_menos_18 = qtde_homens_menos_18 + 1
        elif idade >= 18:
            qtde_homens_mais_18 = qtde_homens_mais_18 + 1
        else:
            print('Valor inválido. Por favor, tente novamente.')

    elif sexo == 'f' or sexo == 'F':
        qtde_mulheres = qtde_mulheres + 1
        if idade < 18 and idade > 0:
            qtde_mulheres_menos_18 = qtde_mulheres_menos_18 + 1
        elif idade >= 18:
            qtde_mulheres_mais_18 = qtde_mulheres_mais_18 + 1
        else:
            print('Valor inválido. Por favor, tente novamente.')
    else:
        print('Input inválido. Por favor, digite (s/n) para confirmar.')
    cadastro = input('Deseja cadastrar (s/n)? ')

print('Quantidade de homens abaixo de 18 anos: ',qtde_homens_menos_18)
print('Quantidade de homens acima de 18 anos: ',qtde_homens_mais_18)
print('Quantidade de mulheres abaixo de 18 anos: ',qtde_mulheres_menos_18)
print('Quantidade de mulheres acima de 18 anos: ',qtde_mulheres_mais_18)

if cadastro == 'n' or cadastro == 'N':
    print()
else:
    print('Input inválido. Por favor, digite (s/n) para confirmar!')
```
