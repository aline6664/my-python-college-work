###1. Faça um algoritmo  que mostre 10 vezes a frase “Bem vindo a Fatec!”. Pode implementar com o comando while ou for.

```
print('Exemplo do while.............\n\n')
# 1º passo - inicialização
contador = 1
# 2º passo - condição/critério de parada
while contador <= 10:
    print(contador, 'Bem vindo a Fatec!')
    contador = contador + 1 # 3º passo - incremento
```

```
print('\n\nExemplo do for.............\n')
# 1º passo - inicialização, 2º passo - condição/critério de parada
# e 3º passo - incremento, numa única linha no for
for contador in range(1,20,2):
    print(contador,'Bem vindo a Fatec!')
print(contador)
```

```
for contador in range(6):
    print(contador,'Bem vindo a Fatec!')
    if contador == 3:
        print(':D :D :D')
```

###6. Faça um algoritmo que receba a idade de 10 pessoas, calcule e exiba a quantidade de pessoas maiores de idade, sendo que a maioridade é obtida após completar 18 anos. Pode implementar com o comando while ou for.


```
#while
q_pessoas_maior18 = 0
q_pessoas = 0

while q_pessoas < 10:
    idade = int(input('Idade: '))
    if idade >= 18:
        q_pessoas_maior18 = q_pessoas_maior18 + 1

    q_pessoas = q_pessoas + 1

print('Quantidade de pessoas maiores de idade: ',q_pessoas_maior18)
```

```
#for
q_pessoas_maior18 = 0

for q_pessoas in range(10):
    idade = int(input('Idade: '))
    if idade >= 18:
        q_pessoas_maior18 = q_pessoas_maior18 + 1

print('Quantidade de pessoas maiores de idade: ',q_pessoas_maior18)
```

###7. Escreva um algoritmo que receba 23 números, calcule e exiba a quantidade de números pares e impares. Pode implementar com o comando while ou for.


```
#while
qtde_par = 0
qtde_impar = 0
contador = 0

while contador < 23:
    numero = int(input('Digite um número: '))
    if numero % 2 == 0: #PAR -> todo par dividido por 2 nunca sobrará resto
        qtde_par = qtde_par + 1
    else:
        qtde_impar = qtde_impar + 1

    contador = contador + 1 
  
print('Quantidade de números pares: ',qtde_par)
print('Quantidade de números ímpares: ',qtde_impar)
```

```
#for
qtde_par = 0
qtde_impar = 0

for contador in range(0,5): #só 5 pra não ficar gigante de novo
    numero = int(input('Digite um número: '))
    if numero % 2 == 0:
        qtde_par = qtde_par + 1
    else:
        qtde_impar = qtde_impar + 1

print('Quantidade de números pares: ',qtde_par)
print('Quantidade de números ímpares: ',qtde_impar)
```

###8. Faça um algoritmo que calcule e exiba o salário reajustado de dez funcionários de acordo com a seguinte regra (pode implementar com o comando while ou for): 
## Salário até 300, reajuste de 50%; 

##Salários maiores que 300, reajuste de 30%.




```
#whil3e
contador = 1
while contador <= 10:
    salario = float(input('Salário do funcionário: '))
    if salario <= 300:
        sal_reajustado = salario + (salario * 50/100)
    else:
        sal_reajustado = salario + (salario * 30/100)
    print('\tSalário reajustado: ',sal_reajustado)

    contador = contador + 1
```

```
#for
for contador in range(1,3):
    salario = float(input('Salário do funcionário: '))
    if salario <= 300:
        sal_reajustado = salario + (salario * 50/100)
    else:
        sal_reajustado = salario + (salario * 30/100)
    print('\tSalário reajustado: ',sal_reajustado)
```

###10. Faça um algoritmo que calcule e informe a média de idades de 5 alunos. Pode implementar com o comando while ou for.

```
#while
q_alunos = 0
soma_idade = 0 
while q_alunos < 5:
    idade = int(input('Idade do aluno: '))
    soma_idade = soma_idade + idade
    q_alunos = q_alunos + 1

media = soma_idade / q_alunos #<- Na média, jamais deixe a divisão dentro da repetição! Apenas a soma irá se repetir.
print('Média de idades =',media)
print(q_alunos)
```

```
#for (observação)
soma_idade = 0
for q_alunos in range(5):
    idade = int(input('Idade do aluno: '))
    soma_idade = soma_idade + idade

media = soma_idade / q_alunos #<- No "for", é necessário prestar atenção nessa parte, pois seu intervalo de contagem é diferente.
#Aqui ele conta apenas 4 alunos, em vez de 5. Logo a média está errada. (18.75 em vez de 15.0)
#Um modo de corrigir seria colocar range(1,6), porém nem em todas situações dá para começar do 1. Ou poderia colocar diretamente soma_idade / 5 (neste caso).
print('Média de idades =',media)
print(q_alunos) #<-4 alunos em vez de 5
```

###11. Faça um algoritmo que conheça 4 preços de produtos, calcule e mostre a média aritmética dos preços. Pode implementar com o comando while ou for.


```
#while
soma_preco = 0
produto = 0
while produto < 4:
    preco = float(input('Informe o preço do produto: '))
    soma_preco = soma_preco + preco
    produto = produto + 1
    
media = soma_preco / produto
print(f'Média aritmética dos preços: {media:.2f}')
```

```
#for
soma_preco = 0
for produto in range(4):
    preco = float(input('Informe o preço do produto: '))
    soma_preco = soma_preco + preco
    
media = soma_preco / produto
print(f'Média aritmética dos preços: {media:.2f}')
```

###12. Faça um programa que receba a idade e a altura de 20 pessoas. Calcule e exiba a média das alturas das pessoas com mais de 20 anos.




```
#while
q_pessoas = 0
soma_altura = 0
while q_pessoas < 4:
    idade = int(input('Idade:'))
    altura = float(input('Altura: '))
    if idade > 20:
        soma_altura = soma_altura + altura
    q_pessoas = q_pessoas + 1

media = soma_altura / q_pessoas
print(f'Média das alturas das pessoas com mais de 20 anos: {media:.2f}')
```

###13. Faça um algoritmo que leia o preço de 20 TV, determine e apresente a média dos preços que possuem valor maior que R$ 1000. Pode implementar com o comando while ou for.

```
#while
q_tv = 0
soma_preco = 0
while q_tv < 20:
    preco_tv = float(input('Informe o preço da TV: '))
    if preco_tv > 1000:
        soma_preco = soma_preco + preco_tv
    q_tv = q_tv + 1

media_preco = soma_preco / q_tv
print('Média dos preços da TV: ',media_preco)
```

```
#for
soma_preco = 0
for q_tv in range(20):
    preco_tv = float(input('Informe o preço da TV: '))
    if preco_tv > 1000:
        soma_preco = soma_preco + preco_tv

media_preco = soma_preco / 20
print('Média dos preços da TV: ',media_preco)
```

###14. Faça um programa que receba a altura e nome  de 5 pessoas. Encontre e apresente a altura da pessoa mais alta e da mais baixa e seus respectivos nomes.

```
#for
for q_pessoas in range(5):
    nome = input('Nome: ')
    altura = float(input('Altura: '))
    if q_pessoas == 0: #<- Essa parte é lida apenas uma vez, quando coloca a primeira variável.
     # Ela é então comparada aos próximos variáveis, e conforme o critério escrito, são substituídos até encontrar o maior ou menor possível.
        alto = altura
        nome_alto = nome
        baixo = altura
        nome_baixo = nome
    elif altura > alto: #A altura da pessoa anterior é comparada ao próximo. Caso atenda os critérios, é substituído.
        alto = altura
        nome_alto = nome
    elif altura < baixo:
        baixo = altura
        nome_baixo = nome

    print('Alto - Nome: ',nome_alto,'\tAltura: ',alto) #Teste para ver os resultados de cada passo
    print('Baixo - Nome: ',nome_baixo,'\tAltura: ',baixo)
```

###15. Faça um programa que receba dez valores inteiros, encontre e apresente o maior e o menor valor.



```
#for
for contador in range(10):
    numero = int(input('Digite um valor inteiro: '))
    if contador == 0: #Mesma coisa do anterior, neste caso, o valor inicial "1" é comparado com o próximo valor "2".
        menor = numero
        maior = numero
    if numero > maior: #O valor "2" por ser maior que "1", substitui-o na parte de "maior valor". E segue assim por diante.
        maior = numero
    if numero < menor: #Por "1" ainda ser o menor de todos, não é substituído e permanece no seu lugar de "menor valor".
        menor = numero

print('Maior valor -',maior)
print('Menor valor -',menor)
```

###17. Faça um programa que calcule e apresente as tabuadas do 3 ao 7. Este exercício utiliza DUAS estruturas de repetição.

```
#while
tabuada = 3
while tabuada <= 7:
    cont = 1 #<- Importante não esquecer de colocar dentro, já que é um "while" que está dentro do outro.
    while cont <= 10:
        print(tabuada,"x",cont,'=',tabuada * cont)
        cont = cont + 1
    print()
    tabuada = tabuada + 1
```

```
#for
for tabuada in range(3,8): #Estrutura de repetição externa. Trabalha 5 vezes e manda ordens pro próximo "for".
    for cont in range(1,11): #Estrutura de repetição interna. Trabalha 50 vezes. Quando termina uma "tabuada", ocorre a repetição no "cont".
        print(tabuada,"x",cont,'=',tabuada * cont) #Onde ocorre a multiplicação. Volta para o começo para realizar o próximo.
    print() #<- Assim que termina "cont", ele aplica esse espaço.
```

```
#while + for?!
#Aqui pode ser misturado uma estrutura repetição com outra. Logo, um "while" pode ter um "for" dentro de si.
tabuada = 3
while tabuada <= 7:
    for cont in range(1,11):
        print(tabuada,"x",cont,'=',tabuada * cont)
    print()
    tabuada = tabuada + 1
```
